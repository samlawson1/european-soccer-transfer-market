```python
def create_league_table(url_base, season_years, league_name):
    result_code = requests.get(url_base)
    soup_code = bs(result_code.content)
    league_table = pd.read_html(str(soup_code))[12]
    league_table['season'] = season_years
    league_table['league_name'] = league_name
    return(league_table)
```

```python
england_1011 = create_league_table(eng_1011, '2010/2011', 'English Premier League')
england_1112 = create_league_table(eng_1112, '2011/2012', 'English Premier League')
england_1213 = create_league_table(eng_1213, '2012/2013', 'English Premier League')
england_1314 = create_league_table(eng_1314, '2013/2014', 'English Premier League')
england_1415 = create_league_table(eng_1415, '2014/2015', 'English Premier League')
england_1516 = create_league_table(eng_1516, '2015/2016', 'English Premier League')
england_1617 = create_league_table(eng_1617, '2016/2017', 'English Premier League')
england_1718 = create_league_table(eng_1718, '2017/2018', 'English Premier League')
england_1819 = create_league_table(eng_1819, '2018/2019', 'English Premier League')
england_tables = pd.concat([england_1011, england_1112, england_1213, england_1314, england_1415, england_1516,
                           england_1617, england_1718, england_1819])
```
